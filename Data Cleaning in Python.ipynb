```python
import pandas as pd
```

```python
df = pd.read_csv('EData.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['InvoiceDate']).dt.date
```

```python
df = df.drop(['InvoiceDate'], axis=1)
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df
```

```python
df1 = df.drop_duplicates(subset='StockCode', keep='first')
df1
```

```python
df2 = df1[['StockCode', 'Description', 'Date']]
df2
```

```python
Orders = df.groupby(by="InvoiceNo")["Quantity"].sum().reset_index()
Orders.head(10)
```

```python
df3 = df.groupby(by="InvoiceNo")["Revenue"].sum().reset_index()
df3['Date'] = df['Date']
```

```python
Orders = df.groupby(by="InvoiceNo")["Quantity"].sum().reset_index()
Orders['Date'] = df['Date']
Orders.head(10)
```

```python
with pd.ExcelWriter('EcommerceData.xlsx') as writer: 
    df.to_excel(writer, sheet_name='Data')
    df2.to_excel(writer, sheet_name='UniqueItems')
    df3.to_excel(writer, sheet_name='InvoiceAmount')
    Orders.to_excel(writer, sheet_name='Orders')
    
```
